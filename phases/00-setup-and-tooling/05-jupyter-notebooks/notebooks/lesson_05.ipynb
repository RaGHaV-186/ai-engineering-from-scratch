## Lesson 05 — Notebook Basics
Exploring a simple dataset with pandas and matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "model": ["Linear", "Random Forest", "Neural Net"],
    "accuracy": [0.72, 0.89, 0.94],
    "train_time_sec": [0.1, 2.3, 45.6]
})
df
```

```python
plt.bar(df["model"], df["accuracy"])
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.show()
```
